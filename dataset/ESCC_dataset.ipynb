```python
geo_path = "./dataset/GEO_clinical_genes.xlsx"
tcga_path = "./dataset/TCGA_clinical_genes.xlsx"
df = pd.read_excel(geo_path)
print(df.shape)
df.head()
```

```python
df.columns = df.columns.str.replace(r'[ :]', '_', regex=True)
df.columns
```

```python
df = df[df['Survival_months'] > 0]
df.shape
```

```python
df = df[(df['Tumor_size'] > 0) & (df['Tumor_size'] < 300)]
df.shape
```

```python
df = df[df['Time_from_diagnosis_to_treatment'] >= 0]
df.shape
```

```python
df["Time_from_diagnosis_to_treatment"] = df["Time_from_diagnosis_to_treatment"] / 30
```

```python
invalid_cases = df[df["Survival_months"] < df["Time_from_diagnosis_to_treatment"]]
print("invalid_cases:", len(invalid_cases))
```
